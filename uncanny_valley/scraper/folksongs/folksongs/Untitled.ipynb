```python
import json
import pandas as pd
import seaborn as sns
```

```python
with open("songs.jl") as f:
    songs = json.loads(f.read())
```

```python
df = pd.DataFrame(songs)
df[df.lyric.str.len() > 0]
```

```python
ser = df.lyric.str.len() > 0

graph = sns.countplot(x=ser)
freq = ser.value_counts() / len(ser)
for f, p in zip(freq,graph.patches):
    height = p.get_height()
    graph.text(p.get_x()+p.get_width()/2., height + 0.1,
        f,ha="center")
```
